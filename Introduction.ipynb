# Introduction

Ce projet s'intéresse à un enjeu majeur de santé publique : les déterminants de la mortalité infantile à l’échelle mondiale. Si la croissance économique, mesurée par le PIB, est souvent considérée comme le moteur principal du développement, son impact réel sur la survie des enfants fait l'objet de nombreux débats. 
En pratique, certains pays à faible revenu, comme l’Ouganda, ont réussi à réduire significativement la mortalité infantile grâce à des investissements ciblés dans les soins de base, la vaccination et l’accès aux services communautaires. À l’inverse, certains pays à revenu élevé ou intermédiare, comme l’Afrique du Sud, présentent encore des disparités régionales et des inefficacités dans l’organisation des services, malgré des dépenses de santé plus élevées, limitant parfois l’impact réel des ressources disponibles.

Dès lors, une question centrale se pose : la richesse d'un pays suffit-elle à garantir une faible mortalité infantile, ou l'organisation et l'intensité de l'offre de soins jouent-elles un rôle plus déterminant ?

Pour répondre à cette question, nous analysons la relation entre niveau de richesse, offre de soins et infantile. Nous partons de l'hypothèse que dans les pays les plus pauvres, de petits investissements dans les services de base peuvent provoquer des progrès spectaculaires, tandis que dans les pays développés, les gains de survie sont plus difficiles à obtenir car le système est déjà performant.

Pour tester ces hypothèses, nous avons constitué une base de données couvrant 25 pays aux niveaux de revenu très variés sur la période 2010-2024. Cette base combine des indicateurs issus de deux sources complémentaires : 

* Les données sanitaires de l'OMS (Organiastion Mondiale de la Santé) (nombre de médecins, taux de vaccination, prévalence de maladies...).

* Indicateurs socio-économiques et démographiques de la Banque Mondiale (PIB par habitant, densité de population, indice de Gini...).


L'originalité de notre approche repose sur deux choix méthodologiques complémentaires. 
D'une part, nous utilisons une Analyse en Composantes Principales (ACP) afin de réduire la dimension d'un ensemble d'une dizaine de variables et synthétiser l'information en trois scores de développement plus lisibles et interprétables. 
D'autre part, nous appliquons un modèle de prédiction XGBoost, permettant de capturer des relations non linéaires entre les variables.
