# Introduction

Ce projet s'intéresse à un enjeu majeur de santé publique : les facteurs qui influencent la mortalité infantile à l’échelle mondiale. Si la croissance économique (le PIB) est souvent considérée comme le moteur principal du développement, son impact réel sur la survie des enfants fait l'objet de nombreux débats. Suffit-il d'être un pays riche pour bien soigner ses enfants, ou est-ce que l'investissement ciblé dans les services publics (vaccination, nombre de médecins) est plus déterminant ?

Notre objectif est d'analyser la relation entre le niveau de richesse, l'offre de soins et la mortalité des jeunes enfants. Nous partons de l'hypothèse que dans les pays les plus pauvres, de petits investissements dans les services de base peuvent provoquer des progrès spectaculaires, tandis que dans les pays développés, les gains de survie sont plus difficiles à obtenir car le système est déjà performant.

Pour tester ces hypothèses, nous avons constitué une base de données couvrant 25 pays aux revenus très variés sur la période 2010-2024. Nous avons regroupé des indicateurs provenant de deux sources complémentaires :

L’OMS : pour les données médicales (nombre de médecins, taux de vaccination, prévalence de maladies).

La Banque Mondiale : pour le contexte économique et démographique (PIB par habitant, densité de population).
